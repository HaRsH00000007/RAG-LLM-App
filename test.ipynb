```python
import os

# Get the value of the environment variable named "api_key"
api_key_value = os.getenv("api_key")

# Check if the value is not None before setting the environment variable
if api_key_value is not None:
    os.environ['api_key'] = api_key_value
else:
    print("The 'api_key' environment variable is not set.")
```

```python
from llama_index import VectorStoreIndex,SimpleDirectoryReader
documents=SimpleDirectoryReader("data").load_data()
```

```python
documents
```

```python
index=VectorStoreIndex.from_documents(documents,show_progress=True)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("What is YOLO")
```

```python
from llama_index.response.pprint_utils import pprint_response
pprint_response(response,show_source=True)
print(response)
```
